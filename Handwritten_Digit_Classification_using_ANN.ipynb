
# **Handwritten Digit Classification using ANN | MNIST Dataset**

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense , Flatten
```

```python
from tensorflow.keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()


```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0])
```

```python
"""Normalize X_train and X_test by dividing by the max value so all features
stay in the same range (0–1). This helps the neural network learn better."""
x_train=x_train/255
x_test=x_test/255
x_train[0]
```

```python
model=Sequential()

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128 , activation='relu'))
model.add(Dense(32 , activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
print("x_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
history=model.fit(x_train , y_train , epochs=25, validation_split=0.2)
```

```python
y_pre=model.predict(x_test)
```

```python
y_pre=y_pre.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , y_pre)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss' , 'val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['accuracy' , 'val_accuracy'])
```
